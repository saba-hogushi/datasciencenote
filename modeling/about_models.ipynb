# ◆機械学習モデルについて
***
機械学習の進め方に関するまとめ。<br>
具体的なモデルについての詳細はここでは触れない。

# ◆モデル作成手順
***
モデルを作成する際の大まかな手順を示す。
1. モデルの種類、パラメータを指定
1. 学習データ、目的変数を与えて学習
1. テストデータを与えて予測
1. モデルの評価(バリデーション)
1. testデータを用いて予測

# ◆バリデーション
***
学習データに対して評価を行っても意味がない。<br>
学習データから一部評価用データを切り分けて置き、評価のために用いる。<br>
このように評価をすることをバリデーションという。<br>

## hold-out法
上記の様に一部のデータを分けておいて評価用に用いることをhold-out法と呼ぶ。<br>
だが、この方法では学習や評価に使用できるデータが少なくなってしまう。<br>
そこでクロスバリデーションを使うことが多い。

## クロスバリデーション
1. 学習データをいくつかに分割する。
1. 分割したものの一つをバリデーションデータとし、その他を学習データとしスコアを求める。
1. 分割した回数だけバリデーションデータを変えて評価を繰り返す。
1. スコアの平均でモデルを評価する。

この方法でモデルの評価がよかった場合どのようにテストデータの予測を行うか。
* 各モデルを保存しておき、それらのモデルの予測の平均値などをとる。
* 改めてすべてのデータに対して学習し、予測を行う。

# ◆用語
***
* **正則化項**<br>
    過学習を防ぐために加えられる項。モデルが複雑になりすぎない。<br>
    <br>
* **アーリーストッピング**<br>
    GBDT、ニューラルネットに組み込まれている機能一定期間スコアが上がらない場合に学習を途中で終える。<br>
    <br>
* **バギング**<br>
    複数のモデルを組み合わせてモデルを作成する方法の一つ。同じ種類のモデルを並列に作成しそれらの出力の平均などを用いて予測。<br>
    データや特徴量を復元抽出で学習する。<br>
    並列化できる<br>
<br>
* **ブースティング**<br>
    複数のモデルを直列につないで出力値を補正しながら学習させていく。<br>
    並列化はできない<br>

# ◆よく使われるモデル
***
* 勾配ブースティング木（GBDT）

* 線形モデル
* その他モデル
